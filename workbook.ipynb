```python
# test spg4_jdfr function 

from jax_sgp4 import sgp4_jdfr, tle2sat, sgp4
from sgp4.api import jday, Satrec
import jax.numpy as jnp

# TLE for the International Space Station (ZARYA)
tle1 = '1 25544U 98067A   25316.22557474  .00019835  00000-0  36009-3 0  9990'
tle2 = '2 25544  51.6334 293.5281 0004132  61.4658 298.6746 15.49560899538121' 

# Create Satrec object
sat = Satrec.twoline2rv(tle1, tle2)
satjax = tle2sat(tle1, tle2)

# # Choose a Julian date to propagate to
# print(jday(2026, 1, 14, 12, 0, 0))  # Example Julian date 14th jan 2026 at 12:00:00 UTC
# jd, fr = jday(2026, 1, 14, 12, 0, 0)

# # the above date gives larger error compared to original sgp4 package since the date is 
# # quite far away and floating point error propagates
# # The below date (one day from epoch) gives similar order of magnitude errors to sgp4 example
# jd, fr = 2460992.5, 0.22557474

# propagate to epoch
jd, fr = sat.jdsatepoch, sat.jdsatepochF 

# jax sgp4_jdfr internal calculations:
year = satjax.epochyr
days, fraction = jnp.divmod(satjax.epochdays, 1.0)
jd_epoch = year * 365 + (year - 1) // 4 + days + 1721044.5
fr_epoch = jnp.round(fraction, 8)
tsince = (jd - jd_epoch) * 1440.0 + (fr - fr_epoch) * 1440.0

# # compare against original sgp4 package jd, fr at epoch
# jd_epoch_ref, fr_epoch_ref = sat.jdsatepoch, sat.jdsatepochF
# print(f'epoch jd jax: {jd_epoch}, epoch fr jax: {fr_epoch}')
# print(f'epoch jd ref: {jd_epoch_ref}, epoch fr ref: {fr_epoch_ref}')

print(f'tsince jax: {tsince}')

jd_jax = jnp.float32(jd)
fr_jax = jnp.float32(fr)

result = sgp4_jdfr(satjax, jd_jax, fr_jax)
r_jax = result[:3]
v_jax = result[3:]

print(f'jax position: {r_jax}')  # position (km)
#print(f'jax velocity: {v_jax}')  # velocity (km/s)

# compare against sgp4 package
e, r_ref, v_ref = sat.sgp4(jd, fr)
print(f'true position: {r_ref}')
print(f'true velocity: {v_ref}')
print(f'difference in position: {r_jax - jnp.array(r_ref)}')
#print(f'difference in velocity: {v_jax - jnp.array(v_ref)}')

# compare against jax with tsince 
result_tsince = sgp4(satjax, 0.0)
r_jax_tsince = result_tsince[:3]
v_jax_tsince = result_tsince[3:]
print(f'difference in position (tsince): {r_jax_tsince - jnp.array(r_ref)} ')
#print(f'difference in velocity (tsince): {v_jax_tsince - jnp.array(v_ref)} ')

print(r_jax.dtype)

# have compared this against the original sgp4 package and the jd, fr converion is works 
# the same, if you have issues then it is probably the sgp4 code itself

print(0.00001 * 60*60*24)
```

```python
# =============================================================================
# Error code validation: compare reference sgp4 error codes against JAX sgp4
# =============================================================================
# These TLEs are from the official SGP4-VER.TLE verification dataset and are
# known to trigger specific error codes when propagated.

from sgp4.api import Satrec
from jax_sgp4 import sgp4, tle2sat
import numpy as np

# Define test cases: (description, tle1, tle2, tsince_values)
error_test_cases = {
    "Error 1 - mean eccentricity out of range (decaying sat, high drag)": {
        "tle1": "1 28872U 05037B   05333.02012661  .25992681  00000-0  24476-3 0  1534",
        "tle2": "2 28872  96.4736 157.9986 0303955 244.0492 110.6523 16.46015938 10708",
        "tsince": [0, 10, 20, 30, 40, 50, 60, 120, 360, 720, 1440],
    },
    "Error 2 - mean motion less than zero (very low mean motion)": {
        "tle1": "1 33334U 78066F   06174.85818871  .00000620  00000-0  10000-3 0  6809",
        "tle2": "2 33334  68.4714 236.1303 5602877 123.7484 302.5767  0.00001000 67521",
        "tsince": [0, 10, 50, 120, 360, 720, 1440],
    },
    "Error 3 - perturbed eccentricity out of range (near-zero incl/ecc)": {
        "tle1": "1 33335U 05008A   06176.46683397 -.00000205  00000-0  10000-3 0  2190",
        "tle2": "2 33335   0.0019 286.9433 0000004  13.7918  55.6504  1.00270176  4891",
        "tsince": [0, 10, 50, 120, 360, 720, 1440],
    },
    "Error 4 - semi-latus rectum negative (near-parabolic ecc=0.995)": {
        "tle1": "1 33333U 05037B   05333.02012661  .25992681  00000-0  24476-3 0  1534",
        "tle2": "2 33333  96.4736 157.9986 9950000 244.0492 110.6523  4.00004038 10708",
        "tsince": [0, 10, 50, 120, 360, 720, 1440],
    },
    "Error 6 - satellite decayed / mrt < 1.0 (extreme drag)": {
        "tle1": "1 29141U 85108AA  06170.26783845  .99999999  00000-0  13519-0 0   718",
        "tle2": "2 29141  82.4288 273.4882 0015848 277.2124  83.9133 15.93343074  6828",
        "tsince": [0, 10, 50, 100, 200, 300, 420],
    },
}

print("SGP4 Error Code Comparison: Reference (C++) vs JAX")
print("=" * 75)

for desc, case in error_test_cases.items():
    print(f"\n{desc}")
    print(f"  TLE1: {case['tle1']}")
    print(f"  TLE2: {case['tle2']}")
    print(f"  {'tsince (min)':<15} {'ref error':<12} {'jax error':<12} {'match'}")
    print(f"  {'-'*55}")
    
    # Create JAX satellite once per TLE
    sat_jax = tle2sat(case["tle1"], case["tle2"])
    
    for t in case["tsince"]:
        # Reference sgp4
        sat_ref = Satrec.twoline2rv(case["tle1"], case["tle2"])
        e_ref, r_ref, v_ref = sat_ref.sgp4_tsince(float(t))
        err_ref = sat_ref.error
        
        # JAX sgp4
        result_jax, err_jax = sgp4(sat_jax, float(t))
        err_jax = int(err_jax)
        
        match = "OK" if err_ref == err_jax else "MISMATCH"
        flag = "" if err_ref == err_jax else " <---"
        print(f"  {t:<15} {err_ref:<12} {err_jax:<12} {match}{flag}")
```
